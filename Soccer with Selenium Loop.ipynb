```python
import time 
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup as bs
import requests
import urllib.request
from selenium import webdriver
import time 

from selenium.common.exceptions import TimeoutException
```

```python
driver = webdriver.Chrome("C:\chromedriver_win32\chromedriver.exe")
url = 'https://sofifa.com/'
driver.get(url)

headers = []
data_by_row = []

USER_AGENT = "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36"
LANGUAGE = "en-US,en;q=0.5"

session = requests.Session()
session.headers['User-Agent'] = USER_AGENT
session.headers['Accept-Language'] = LANGUAGE
session.headers['Content-Language'] = LANGUAGE
html = requests.get(driver.current_url).text

soup = bs(html, 'html.parser')
    
table = soup.find('table')
table_rows = table.find_all('tr')
table_headers = table.find_all('th')
    
#Extract Headers
for th in table_headers:
    headers.append(th.text.replace('\n', ' ').strip())
```

```python
next_button_xpath = "//*[contains(text(), 'Next')]"

while True:
    USER_AGENT = "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36"
    LANGUAGE = "en-US,en;q=0.5"

    session = requests.Session()
    session.headers['User-Agent'] = USER_AGENT
    session.headers['Accept-Language'] = LANGUAGE
    session.headers['Content-Language'] = LANGUAGE
    html = requests.get(driver.current_url).text

    soup = bs(html, 'html.parser')
    
    table = soup.find('table')
    table_rows = table.find_all('tr') 
    
    #Extract Rows
    for tr in table_rows:
        td = tr.find_all('td')
        row = [i.text.replace('\n', ' ').strip() for i in td]
        data_by_row.append(row)
    
    if len(driver.find_elements_by_xpath(next_button_xpath)) < 1:
        df = pd.DataFrame(data_by_row, columns= headers)
        break

    driver.find_element_by_xpath(next_button_xpath).click() 
```

```python
table = soup.find('table')
table_rows = table.find_all('tr')
table_headers = table.find_all('th')
```

```python
headers = []
for th in table_headers:
    headers.append(th.text.replace('\n', ' ').strip())
```

```python
data_by_row = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text.replace('\n', ' ').strip() for i in td]
    data_by_row.append(row)
```

```python
df = pd.DataFrame(data_by_row, columns= headers)
```
